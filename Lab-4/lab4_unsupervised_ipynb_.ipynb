Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
n_clusters = n_classes

def kmeans_custom(X, n_clusters, n_init=100, random_state=None, max_iter=300):
    rng = np.random.RandomState(random_state)
    best_inertia = None
    best_labels = None
    best_centers = None

    for _ in range(n_init):
        centers = X[rng.choice(X.shape[0], n_clusters, replace=False)]

        for _ in range(max_iter):
            dists = ((X[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
            labels = dists.argmin(axis=1)

            new_centers = np.array(
                [X[labels == j].mean(axis=0) if np.any(labels == j) else centers[j]
                 for j in range(n_clusters)]
            )

            if np.allclose(centers, new_centers):
                break
            centers = new_centers

        inertia = ((X - centers[labels]) ** 2).sum()
        if best_inertia is None or inertia < best_inertia:
            best_inertia = inertia
            best_labels = labels
            best_centers = centers

    return best_labels, best_centers

labels_custom, centers_custom = kmeans_custom(
    X_pca, n_clusters=n_clusters, n_init=100, random_state=RANDOM_STATE
)

kmeans = KMeans(n_clusters=n_clusters, n_init=100, random_state=RANDOM_STATE)
kmeans.fit(X_pca)
labels_sklearn = kmeans.labels_
````

````python
inertia = []
for k in tqdm_notebook(range(1, n_classes + 1)):
    kmeans_k = KMeans(n_clusters=k, n_init=100, random_state=RANDOM_STATE)
    kmeans_k.fit(X_pca)
    inertia.append(kmeans_k.inertia_)

plt.plot(range(1, n_classes + 1), inertia)
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.show()
````
